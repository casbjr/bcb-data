```python
import pandas as pd
import requests
```

```python
# 1. Carrega o Cadastro de Março de 2026
print("Carregando cadastro do Banco Central...")
url_cad = "https://olinda.bcb.gov.br/olinda/servico/IFDATA/versao/v1/odata/IfDataCadastro(AnoMes=@AnoMes)?@AnoMes=202512&$format=json"
r_cad = requests.get(url_cad).json()
df_cad = pd.DataFrame(r_cad['value'])

# Filtra as instituições usando seus termos de busca
termos = ["PORTO SEGURO", "PORTO BANK", "BANCO PAN", "BANCO VOTORANTIM", "BV FINANCEIRA", "BANCO INTER", "C6 BANK", "BANCO C6", "ITAÚ UNIBANCO", "ITAU UNIBANCO", "BRADESCO", "SANTANDER", "BTG PACTUAL", "NU PAGAMENTOS", "NU FINANCEIRA"]
padrao = "|".join(termos)
df_cad_filtrado = df_cad[df_cad["NomeInstituicao"].str.contains(padrao, case=False, na=False)]

print("\n--- AMOSTRA DO SEU CADASTRO FILTRADO ---")
print(df_cad_filtrado[['CodInst', 'NomeInstituicao', 'CodConglomeradoPrudencial', 'CnpjInstituicaoLider']].tail(10).to_string())
```

```python
# 2. Carrega os Valores do Relatório 11 (Cartões)
print("\nCarregando valores do Relatório 11...")
url_val = "https://olinda.bcb.gov.br/olinda/servico/IFDATA/versao/v1/odata/IfDataValores(AnoMes=@AnoMes,TipoInstituicao=@TipoInstituicao,Relatorio=@Relatorio)?@AnoMes=202512&@TipoInstituicao=1&@Relatorio='11'&$format=json"

r_val = requests.get(url_val).json()
df_val = pd.DataFrame(r_val['value'])

print("\n--- AMOSTRA DOS VALORES DO RELATÓRIO DO BCB ---")
#print(df_val[['CodInst', 'NomeRelatorio', 'NumeroRelatorio', 'NomeColuna', 'DescricaoColuna', 'Saldo']].tail(10).to_string())
df_val.tail(10)
```

```python
# 1. Certifique-se de que a função de normalização está carregada nesta célula
def normalizar_codigo(val) -> str:
    import pandas as pd
    if pd.isna(val):
        return ""
    s = str(val).strip()
    if s.endswith('.0'):
        s = s[:-2]
    s_numerico = "".join(filter(str.isdigit, s))
    if s_numerico:
        return s_numerico.zfill(8)
    return ""

# 2. Prepara os DataFrames para o Join
df_cad_full = df_cad.copy()
df_cad_full['CodInst_limpo'] = df_cad_full['CodInst'].apply(normalizar_codigo)

if 'CodConglomeradoPrudencial' in df_cad_full.columns:
    df_cad_full['CodPrud_limpo'] = df_cad_full['CodConglomeradoPrudencial'].apply(normalizar_codigo)
else:
    df_cad_full['CodPrud_limpo'] = ""

df_val_full = df_val.copy()
df_val_full['CodInst_limpo'] = df_val_full['CodInst'].apply(normalizar_codigo)

# Match 1: Cruzando CodInst do Relatório com o CodInst do Cadastro (CNPJ base)
match_cnpj = df_val_full.merge(df_cad_full, on='CodInst_limpo', how='inner')
bancos_cnpj = match_cnpj['NomeInstituicao'].unique()

# Match 2: Cruzando CodInst do Relatório com o Conglomerado Prudencial do Cadastro
match_prud = df_val_full.merge(df_cad_full, left_on='CodInst_limpo', right_on='CodPrud_limpo', how='inner')
bancos_prud = match_prud['NomeInstituicao_y'].unique() if 'NomeInstituicao_y' in match_prud.columns else match_prud['NomeInstituicao'].unique()

print("--- DIAGNÓSTICO DE MATCH ---")
print(f"Total de casamentos via CNPJ Base (CodInst): {len(match_cnpj)}")
print(f"Amostra de bancos via CNPJ: {list(bancos_cnpj)[:15]}\n")

print(f"Total de casamentos via Conglomerado Prudencial: {len(match_prud)}")
print(f"Amostra de bancos via Prudencial: {list(bancos_prud)[:15]}\n")

# Verifica se algum dos seus alvos aparece em alguma das duas listas de match
todos_bancos_no_relatorio = set(match_cnpj['NomeInstituicao'].dropna().unique())
if 'NomeInstituicao_y' in match_prud.columns:
    todos_bancos_no_relatorio |= set(match_prud['NomeInstituicao_y'].dropna().unique())
else:
    todos_bancos_no_relatorio |= set(match_prud['NomeInstituicao'].dropna().unique())

alvos_presentes = [b for b in todos_bancos_no_relatorio if any(t in b.upper() for t in ["BRADESCO", "ITAÚ", "ITAU", "NU ", "C6", "PAN", "INTER", "VOTORANTIM"])]
print("--- TARGETS ENCONTRADOS ---")
print("Bancos do seu target presentes no relatório 11 de 202512:")
print(alvos_presentes)
```

```python
import pandas as pd
import requests
import json

# 1. Definindo a função diretamente na célula para evitar NameError
def normalizar_codigo(val) -> str:
    if pd.isna(val):
        return ""
    s = str(val).strip()
    if s.endswith('.0'):
        s = s[:-2]
    s_numerico = "".join(filter(str.isdigit, s))
    if s_numerico:
        return s_numerico.zfill(8)
    return ""

def obter_dados_paginados(url_inicial) -> list:
    """Faz requisições consecutivas seguindo o link @odata.nextLink se houver."""
    resultados = []
    url_atual = url_inicial
    headers = {"User-Agent": "Mozilla/5.0"}
    
    pagina = 1
    while url_atual:
        print(f"  Baixando página {pagina}...")
        resp = requests.get(url_atual, headers=headers, timeout=60)
        
        if resp.status_code != 200:
            raise RuntimeError(f"Erro na página {pagina}: {resp.status_code}")
            
        corpo = resp.text.strip()
        if corpo.startswith("/*") and corpo.endswith("*/"):
            corpo = corpo[2:-2].strip()
            
        # Parser seguro usando o JSON carregado da resposta limpa
        dados = json.loads(corpo)
        if isinstance(dados, dict):
            resultados.extend(dados.get("value", []))
            url_atual = dados.get("@odata.nextLink")
        else:
            if isinstance(dados, list):
                resultados.extend(dados)
            break
        pagina += 1
    return resultados

# 1. Carrega o Cadastro de Dezembro de 2025
print("Carregando cadastro do Banco Central (202512)...")
url_cad = "https://olinda.bcb.gov.br/olinda/servico/IFDATA/versao/v1/odata/IfDataCadastro(AnoMes=@AnoMes)?@AnoMes=202512&$format=json"
r_cad = requests.get(url_cad).json()
df_cad = pd.DataFrame(r_cad['value'])

# Filtra as instituições usando seus termos de busca
termos = ["PORTO SEGURO", "PORTO BANK", "BANCO PAN", "BANCO VOTORANTIM", "BV FINANCEIRA", "BANCO INTER", "C6 BANK", "BANCO C6", "ITAÚ UNIBANCO", "ITAU UNIBANCO", "BRADESCO", "SANTANDER", "BTG PACTUAL", "NU PAGAMENTOS", "NU FINANCEIRA"]
padrao = "|".join(termos)
df_cad_filtrado = df_cad[df_cad["NomeInstituicao"].str.contains(padrao, case=False, na=False)].copy()

# 2. Carrega os Valores com Paginação (Tipo 1 - onde vivem em 2025/2026)
print("\nCarregando valores do Relatório 11 com Paginação OData...")
url_val_base = "https://olinda.bcb.gov.br/olinda/servico/IFDATA/versao/v1/odata/IfDataValores(AnoMes=@AnoMes,TipoInstituicao=@TipoInstituicao,Relatorio=@Relatorio)?@AnoMes=202512&@TipoInstituicao=1&@Relatorio='11'&$format=json"

valores_completos = obter_dados_paginados(url_val_base)
df_val = pd.DataFrame(valores_completos)

# 3. Valida a interseção de forma dinâmica
col_cod_cad = 'CodInst' if 'CodInst' in df_cad_filtrado.columns else df_cad_filtrado.columns[0]
col_cod_val = 'CodInst' if 'CodInst' in df_val.columns else df_val.columns[0]

# Normalizando os códigos do cadastro
codigos_cad = set(df_cad_filtrado[col_cod_cad].apply(normalizar_codigo).dropna().unique())
if 'CodConglomeradoPrudencial' in df_cad_filtrado.columns:
    codigos_cad |= set(df_cad_filtrado['CodConglomeradoPrudencial'].apply(normalizar_codigo).dropna().unique())

df_val['CodInst_limpo'] = df_val[col_cod_val].apply(normalizar_codigo)
codigos_val = set(df_val['CodInst_limpo'].unique())

intersecao = codigos_cad & codigos_val

print("\n--- RESULTADO DO CASAMENTO DE CHAVES COM PAGINAÇÃO ---")
print(f"Total de códigos mapeados no cadastro: {len(codigos_cad)}")
print(f"Total de códigos retornados no relatório completo: {len(codigos_val)}")
print(f"Interseção direta (Bancos Encontrados): {intersecao}")
```
